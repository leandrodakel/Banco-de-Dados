```python
import pandas as pd

# Lendo uma tabela da internet e transformando-a em um DataFrame

url = 'https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/'

dfs = pd.read_html(url)

print(type(dfs))

print(len(dfs))

# Aqui, teremos como resultado nosso DataFrame com uma lista que cont√©m apenas uma tabela.
```

```python
df_bancos = dfs[0]

print(df_bancos.shape)

print(df_bancos.dtypes)

df_bancos.head()
```
